```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
base_path = '../Data/wassa_data/'
```

```python
def plot_histogram(emotion, directory, ax, ending):
    file_name = f'{base_path}{directory}/{emotion}{ending}' 

    df = pd.read_csv(file_name, delimiter='\t')
    sns.histplot(ax = ax, data=df, x = 'score')
    ax.set_title(emotion)
```

```python
def plot_histogram_len(emotion, directory, ax, ending):
    file_name = f'{base_path}{directory}/{emotion}{ending}' 

    df = pd.read_csv(file_name, delimiter='\t')
    df['len'] = df['tweet'].apply(lambda x: len(x))
    sns.histplot(ax = ax, data=df, x = 'score')
    ax.set_title(emotion)
```

```python
emotions = ['anger', 'joy', 'sadness', 'fear']
```

## Training Data

```python
directory = 'training'
ending = '-ratings-0to1.train.txt'
```

```python
fig, axes = plt.subplots(2, 2, figsize=(24,12))
for i in range(len(emotions)):
    plot_histogram(emotions[i], directory, axes[i//2][i%2], ending)
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(24,12))
for i in range(len(emotions)):
    plot_histogram(emotions[i], directory, axes[i//2][i%2], ending)
plt.show()
```

## Validation Data

```python
directory = 'validation'
ending = '-ratings-0to1.dev.gold.txt'

fig, axes = plt.subplots(2, 2, figsize=(24,12))
for i in range(len(emotions)):
    plot_histogram_len(emotions[i], directory, axes[i//2][i%2], ending)
plt.show()
```

## Testing Data

```python
directory = 'testing'
ending = '-ratings-0to1.test.target.txt'

fig, axes = plt.subplots(2, 2, figsize=(24,12))
for i in range(len(emotions)):
    plot_histogram(emotions[i], directory, axes[i//2][i%2], ending)
plt.show()
```
