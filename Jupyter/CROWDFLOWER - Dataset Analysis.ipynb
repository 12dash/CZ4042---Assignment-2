```python
import pandas as pd
import seaborn as sns

import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../Data/text_emotion.csv')
df.head()
```

```python
print('Number of tweets : ', df.size)
```

```python
sns.set(font_scale=2)
plt.figure(1, figsize=(24,12))
sns.countplot(x='sentiment', data=df)
plt.show()
```
